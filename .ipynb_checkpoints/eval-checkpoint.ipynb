```python
# Cell 1: Imports


import

os
,

json
,

random
,

time


import

numpy

as

np


import

cv2


import

matplotlib.pyplot

as

plt
```

```python
# Cell 2: Paths + hyperparameters (EDIT THESE)


SPLIT_PATH

=

"splits/lfw_1N_split.json"


G_IDS_PATH

=

"cache/gallery_ids.npy"


G_EMB_PATH

=

"cache/G.npy"



# Point this to your saved baseline saliency map (112x112)


SAL_PATH

=

"results/rise_alignedchip_baseline/rise_baseline_cosine_N1000_s8_p0.5_seed123_saliency_norm.npy"



# Evaluation settings


STEPS

=

50

# number of points on curve


INSERTION_BASELINE

=

"blur"

# "blur" or "black"


BLUR_KSIZE

=

11

# odd integer, used if baseline="blur"



SEED

=

123
```

```python
# Cell 3: Load split + gallery embeddings


with

open
(
SPLIT_PATH
,

"r"
)

as

f
:


split

=

json
.
load
(
f
)



gallery

=

split
[
"gallery"
]


probes

=

split
[
"probes"
]



gallery_ids

=

np
.
load
(
G_IDS_PATH
,

allow_pickle
=
True
)
.
tolist
()


G

=

np
.
load
(
G_EMB_PATH
)
.
astype
(
np
.
float32
)



id_to_index

=

{
pid
:

i

for

i
,

pid

in

enumerate
(
gallery_ids
)}



def

get_gallery_emb
(
person_id
:

str
)

->

np
.
ndarray
:


e

=

G
[
id_to_index
[
person_id
]]
.
astype
(
np
.
float32
)


return

e

/

(
np
.
linalg
.
norm
(
e
)

+

1e-12
)



print
(
"Gallery embeddings:"
,

G
.
shape
,

"| identities:"
,

len
(
gallery_ids
))
```

```python
# Cell 4: Setup InsightFace (recognition-only embedding from aligned chip)


from

insightface.app

import

FaceAnalysis



app

=

FaceAnalysis
(
name
=
"buffalo_l"
,

providers
=
[
"CUDAExecutionProvider"
,

"CPUExecutionProvider"
])


app
.
prepare
(
ctx_id
=
0
,

det_size
=
(
640
,

640
))


rec

=

app
.
models
[
"recognition"
]



def

get_embedding_from_chip
(
chip_bgr
:

np
.
ndarray
)

->

np
.
ndarray
:



"""chip_bgr: (112,112,3) BGR aligned face crop -> (512,) L2-normalized"""


assert

chip_bgr
.
shape
[:
2
]

==

(
112
,

112
)

and

chip_bgr
.
shape
[
2
]

==

3
,

f
"Expected (112,112,3), got
{
chip_bgr
.
shape
}
"


feat

=

rec
.
get_feat
(
chip_bgr
)


feat

=

np
.
asarray
(
feat
)
.
reshape
(
-
1
)
.
astype
(
np
.
float32
)


feat

=

feat

/

(
np
.
linalg
.
norm
(
feat
)

+

1e-12
)


return

feat



def

cosine_score
(
chip_bgr
:

np
.
ndarray
,

gallery_emb
:

np
.
ndarray
)

->

float
:


e

=

get_embedding_from_chip
(
chip_bgr
)


return

float
(
np
.
dot
(
e
,

gallery_emb
))



print
(
"InsightFace ready. Rec model:"
,

type
(
rec
))
```

```python
# Cell 5: Pick a probe (EDIT if you want a specific identity/image)


random
.
seed
(
SEED
)



valid_probe_ids

=

[
pid

for

pid

in

probes
.
keys
()

if

pid

in

gallery

and

len
(
probes
[
pid
])

>

0
]


true_id

=

random
.
choice
(
valid_probe_ids
)


probe_path

=

random
.
choice
(
probes
[
true_id
])



print
(
"True ID:"
,

true_id
)


print
(
"Probe path:"
,

probe_path
)


print
(
"Gallery path:"
,

gallery
[
true_id
])



img

=

cv2
.
imread
(
probe_path
)


assert

img

is

not

None
,

f
"Could not read
{
probe_path
}
"



faces

=

app
.
get
(
img
)


assert

len
(
faces
)

>

0
,

"No face detected on the probe."



face

=

max
(
faces
,

key
=
lambda

f
:

(
f
.
bbox
[
2
]
-
f
.
bbox
[
0
])

*

(
f
.
bbox
[
3
]
-
f
.
bbox
[
1
]))


assert

hasattr
(
face
,

"kps"
)

and

face
.
kps

is

not

None
,

"No 5-point landmarks (kps) found."



src

=

face
.
kps
.
astype
(
np
.
float32
)


dst

=

np
.
array
([


[
38.2946
,

51.6963
],


[
73.5318
,

51.5014
],


[
56.0252
,

71.7366
],


[
41.5493
,

92.3655
],


[
70.7299
,

92.2041
],


],

dtype
=
np
.
float32
)



M
,

_

=

cv2
.
estimateAffinePartial2D
(
src
,

dst
,

method
=
cv2
.
LMEDS
)


assert

M

is

not

None
,

"Failed to estimate alignment transform."



chip_bgr

=

cv2
.
warpAffine
(
img
,

M
,

(
112
,

112
),

flags
=
cv2
.
INTER_LINEAR
,

borderValue
=
0
)


chip_rgb

=

cv2
.
cvtColor
(
chip_bgr
,

cv2
.
COLOR_BGR2RGB
)



print
(
"Aligned chip:"
,

chip_bgr
.
shape
)
```

```python
# Cell 6: Load saliency map (must be 112x112 for aligned-chip evaluation)


sal

=

np
.
load
(
SAL_PATH
)
.
astype
(
np
.
float32
)


assert

sal
.
shape

==

(
112
,

112
),

f
"Expected saliency shape (112,112), got
{
sal
.
shape
}
"


print
(
"Loaded saliency:"
,

SAL_PATH
)



# Sanity: visualize overlay on aligned chip


plt
.
figure
(
figsize
=
(
4
,
4
))


plt
.
imshow
(
chip_rgb
)


plt
.
imshow
(
sal
,

alpha
=
0.5
,

vmin
=
0
,

vmax
=
1
)


plt
.
axis
(
"off"
)


plt
.
title
(
"Saliency overlay (aligned 112x112 chip)"
)


plt
.
show
()
```

```python
# Cell 7: Build insertion baseline image


def

make_insertion_baseline
(
chip_bgr
:

np
.
ndarray
,

mode
:

str

=

"blur"
,

ksize
:

int

=

11
)

->

np
.
ndarray
:


if

mode

==

"black"
:


return

np
.
zeros_like
(
chip_bgr
,

dtype
=
np
.
uint8
)


if

mode

==

"blur"
:


k

=

int
(
ksize
)


if

k

%

2

==

0
:


k

+=

1


return

cv2
.
GaussianBlur
(
chip_bgr
,

(
k
,

k
),

0
)


raise

ValueError
(
"mode must be 'blur' or 'black'"
)



baseline_bgr

=

make_insertion_baseline
(
chip_bgr
,

mode
=
INSERTION_BASELINE
,

ksize
=
BLUR_KSIZE
)
```

```python
# Cell 8: Insertion / Deletion evaluation


def

insertion_deletion_curves
(


chip_bgr
:

np
.
ndarray
,


saliency
:

np
.
ndarray
,


gallery_emb
:

np
.
ndarray
,


steps
:

int

=

50
,


insertion_baseline_bgr
:

np
.
ndarray

|

None

=

None
,


):



"""


    Returns:


      frac: (steps+1,) fraction of pixels modified (0..1)


      deletion_scores: (steps+1,)


      insertion_scores: (steps+1,)


    """


H
,

W

=

saliency
.
shape


n_pix

=

H

*

W


order

=

np
.
argsort
(
saliency
.
reshape
(
-
1
))[::
-
1
]

# most important first



# deletion starts from original


del_img

=

chip_bgr
.
copy
()



# insertion starts from baseline (blur or black)


if

insertion_baseline_bgr

is

None
:


ins_img

=

np
.
zeros_like
(
chip_bgr
,

dtype
=
np
.
uint8
)


else
:


ins_img

=

insertion_baseline_bgr
.
copy
()



# We will modify pixels progressively in chunks


# Include step 0 (no pixels changed) and step=steps (all pixels changed)


frac

=

np
.
linspace
(
0.0
,

1.0
,

steps

+

1
)


deletion_scores

=

[]


insertion_scores

=

[]



# Score at step 0


deletion_scores
.
append
(
cosine_score
(
del_img
,

gallery_emb
))


insertion_scores
.
append
(
cosine_score
(
ins_img
,

gallery_emb
))



for

t

in

range
(
1
,

steps

+

1
):


k

=

int
(
round
(
frac
[
t
]

*

n_pix
))


idx

=

order
[:
k
]


ys

=

idx

//

W


xs

=

idx

%

W



# deletion: zero out top-k pixels


del_img
[
ys
,

xs
]

=

0



# insertion: copy in top-k pixels from original


ins_img
[
ys
,

xs
]

=

chip_bgr
[
ys
,

xs
]



deletion_scores
.
append
(
cosine_score
(
del_img
,

gallery_emb
))


insertion_scores
.
append
(
cosine_score
(
ins_img
,

gallery_emb
))



return

frac
,

np
.
array
(
deletion_scores
,

dtype
=
np
.
float32
),

np
.
array
(
insertion_scores
,

dtype
=
np
.
float32
)



t_y

=

get_gallery_emb
(
true_id
)



frac
,

del_scores
,

ins_scores

=

insertion_deletion_curves
(


chip_bgr
=
chip_bgr
,


saliency
=
sal
,


gallery_emb
=
t_y
,


steps
=
STEPS
,


insertion_baseline_bgr
=
baseline_bgr
,


)



# AUC via trapezoid rule


del_auc

=

float
(
np
.
trapz
(
del_scores
,

frac
))


ins_auc

=

float
(
np
.
trapz
(
ins_scores
,

frac
))



print
(
"Deletion AUC (lower better):"
,

del_auc
)


print
(
"Insertion AUC (higher better):"
,

ins_auc
)



print
(
"Score at start (clean):"
,

float
(
del_scores
[
0
]))


print
(
"Score at end (all deleted):"
,

float
(
del_scores
[
-
1
]))


print
(
"Score at start (baseline):"
,

float
(
ins_scores
[
0
]))


print
(
"Score at end (all inserted):"
,

float
(
ins_scores
[
-
1
]))
```

```python
# Cell 9: Plot curves


plt
.
figure
(
figsize
=
(
6
,
4
))


plt
.
plot
(
frac
,

del_scores
,

label
=
"Deletion"
)


plt
.
plot
(
frac
,

ins_scores
,

label
=
"Insertion"
)


plt
.
xlabel
(
"Fraction of pixels modified"
)


plt
.
ylabel
(
"Cosine similarity to true ID embedding"
)


plt
.
title
(
"Insertion/Deletion Curves (Aligned 112×112 chip)"
)


plt
.
legend
()


plt
.
grid
(
True
)


plt
.
show
()



print
(
"Deletion AUC:"
,

del_auc
)


print
(
"Insertion AUC:"
,

ins_auc
)
```

```python
# Cell 10 (Optional): Visualize intermediate deletion/insertion images


# Pick a few fractions to visualize


fractions_to_show

=

[
0.0
,

0.1
,

0.25
,

0.5
,

0.75
,

1.0
]



H
,

W

=

sal
.
shape


n_pix

=

H

*

W


order

=

np
.
argsort
(
sal
.
reshape
(
-
1
))[::
-
1
]



def

render_state
(
frac_val
:

float
):


k

=

int
(
round
(
frac_val

*

n_pix
))


idx

=

order
[:
k
]


ys

=

idx

//

W


xs

=

idx

%

W



del_img

=

chip_bgr
.
copy
()


del_img
[
ys
,

xs
]

=

0



ins_img

=

baseline_bgr
.
copy
()


ins_img
[
ys
,

xs
]

=

chip_bgr
[
ys
,

xs
]



return

del_img
,

ins_img



for

fval

in

fractions_to_show
:


del_img
,

ins_img

=

render_state
(
fval
)



del_rgb

=

cv2
.
cvtColor
(
del_img
,

cv2
.
COLOR_BGR2RGB
)


ins_rgb

=

cv2
.
cvtColor
(
ins_img
,

cv2
.
COLOR_BGR2RGB
)



plt
.
figure
(
figsize
=
(
8
,
3
))


plt
.
subplot
(
1
,
2
,
1
)


plt
.
imshow
(
del_rgb
)


plt
.
axis
(
"off"
)


plt
.
title
(
f
"Deletion @
{
fval
:
.2f
}
"
)



plt
.
subplot
(
1
,
2
,
2
)


plt
.
imshow
(
ins_rgb
)


plt
.
axis
(
"off"
)


plt
.
title
(
f
"Insertion @
{
fval
:
.2f
}
"
)


plt
.
show
()
```
